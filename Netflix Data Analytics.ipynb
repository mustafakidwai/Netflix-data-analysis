### Q3. How many Movies and Tv shows are in the dataset?

```python
data.head(2)
```

```python
data.groupby('Category').Category.count()
```

```python
sns.countplot(data['Category'])
```

```python
movie_rows = data.loc[data["Category"] == "Movie"]
tv_rows = data.loc[data["Category"] == "TV Show"]

# Calculate percentages of movies and tv shows:
movies_per = round(movie_rows.shape[0] / data["Category"].shape[0] * 100, 2)
tvshows_per = round(tv_rows.shape[0] / data["Category"].shape[0] * 100, 2)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10, 6))

plt.pie([movies_per, tvshows_per], labels=["Movie", "TV Show"], shadow=True, startangle=0, autopct ='% 1.1f %%', explode=(0, 0.1), colors=sns.color_palette("RdGy", n_colors=2))
plt.axis('equal')

plt.title('Percentage of movies and TV shows\n', fontsize=15, fontweight='bold')
plt.legend(["Movie", "TV Show"], loc=0);
```

### Q14: Top 10 producing countries

```python
#group the number of titles produced per country and then will plot two figures
most_titles_produced = data.groupby('Country').count().sort_values('Title',ascending=False).head(10)

most_titles_produced.reset_index(inplace=True)
most_titles_produced
```

```python
from matplotlib import gridspec

fig = plt.figure(figsize=(20, 6))
gs = gridspec.GridSpec(nrows=1, ncols=2, height_ratios=[6], width_ratios=[10, 5])

ax = plt.subplot(gs[0])

sns.barplot(x=most_titles_produced.Country, y=most_titles_produced.Title, data=most_titles_produced, palette="RdGy")
#hue colors: Movies and TV

ax.set_xticklabels(most_titles_produced.Country, fontweight='bold', rotation='90')
ax.set_title('Top 10 producing countries\n', fontsize=15, fontweight='bold');
```
